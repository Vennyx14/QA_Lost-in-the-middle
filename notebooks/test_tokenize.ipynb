# Test vncorenlp

```python
import py_vncorenlp

py_vncorenlp.__file__
```

```python
vncorenlp_model_path = '../model'

py_vncorenlp.download_model(save_dir=vncorenlp_model_path)
vncorenlp_model = py_vncorenlp.VnCoreNLP(save_dir=vncorenlp_model_path)
```

```python
def vncorenlp_tokenize(text):
    output = []
    annotated_text = vncorenlp_model.annotate_text(text)[0]

    for sent in annotated_text:
        output.append(
            sent['wordForm'].replace('_', ' '),
        )

    return output
```

```python
sentence = "Với cộng đồng người Bách Việt trước đây, việc thuần hóa mèo cũng có thể theo cách thức như vậy."

vncorenlp_tokenize(sentence)
```

# Test word2vec

```python
from gensim.models import KeyedVectors

# Load model
path_to_model = "../word2vec/word2vec_vi_words_100dims.txt"
model = KeyedVectors.load_word2vec_format(path_to_model, binary=False)
```

```python
# Check similarity
result = model.similarity('học_sinh', 'sinh_viên')

f"Độ tương đồng: {result}"
```

```python
model.similarity('vua', 'hoàng_hậu')
```

```python
model.most_similar(positive=['hoàng_hậu', 'nam'], negative=['nữ'])
```

```python
model.most_similar('Nhật_bản')
```

```python
result = model.most_similar(positive=['Đức', 'Mussolini'], negative=['Hitler'])

result
```

```python
def preprocess_tokens(tokens, wv_model) :
    new_tokens = []
    
    for token in tokens :
        if token in wv_model.key_to_index :
            new_tokens.append(token)
        else :
            new_token = '_'.join(token.split(' '))

            if new_token in wv_model.key_to_index :
                new_tokens.append(new_token)
            else :
                print(f"{new_token} not in corpus")

    return new_tokens
```

```python
def embedding(sentence) :
    tokens = vncorenlp_tokenize(sentence)
    tokens = preprocess_tokens(tokens, model)

    return model[tokens]
```
