```python
from datasets import load_dataset
import json
import pandas as pd
```

# Download taidng/UIT-ViQuAD2.0

```python
ds = load_dataset("taidng/UIT-ViQuAD2.0")

len(ds)
```

```python
output_dir = '../data/raw/UIT-ViQuAD2.0/'

for split in ('train', 'validation', 'test') :
    data = ds[split]

    output_filename = output_dir + split + '.jsonl'
    with open(output_filename, 'w') as jsonl_file :
        for d in data :
            json_string = json.dumps(d, ensure_ascii=False)
            jsonl_file.write(json_string + '\n')

    print("Successfully write to " + output_filename)
```

# Download anti-ai-community/tydiqa-goldp-vietnamese

```python
# Load corpuses and queries

corpuses = load_dataset("anti-ai-community/tydiqa-goldp-vietnamese", "corpus")
queries = load_dataset("anti-ai-community/tydiqa-goldp-vietnamese", "query")

# Data only contains test split
corpuses, queries = corpuses['test'], queries['test']

len(corpuses), len(queries)
```

```python
corpuses[0]
```

```python
base_output_dir = '../data/raw/tydiqa-goldp-vietnamese/'

corpus_output_filename = base_output_dir + 'corpuses' + '.jsonl'
with open(corpus_output_filename, 'w') as jsonl_file :
    for c in corpuses :
        json_string = json.dumps(c, ensure_ascii=False)
        jsonl_file.write(json_string + '\n')

print("Successfully write to " + corpus_output_filename)
```

```python
queries_output_filename = base_output_dir + 'queries' + '.jsonl'
with open(queries_output_filename, 'w') as jsonl_file :
    for q in queries :
        json_string = json.dumps(q, ensure_ascii=False)
        jsonl_file.write(json_string + '\n')

print("Successfully write to " + queries_output_filename)
```

# Download mmarco-EnVi-subdataset

```python
# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("tiennv/mmarco-EnVi-subdataset")
```

```python
ds = pd.DataFrame(ds['train'])

ds
```

```python
english_words = ['of', 'an', 'the']
english_pattern = '|'.join(english_words)

vietnamese_ds = ds[~ds['positive'].str.contains(english_pattern)]

vietnamese_ds[:8]
```

```python
base_output_dir = '../data/raw/mmarco-EnVi-subdataset/'

json_vietnamese_df = vietnamese_ds.to_json()

output_filename = base_output_dir + 'train' + '.jsonl'
with open(output_filename, 'w') as jsonl_file :
    for line in vietnamese_ds :
        json_string = json.dumps(line, ensure_ascii=False)
        jsonl_file.write(json_string + '\n')

print("Successfully write to " + output_filename)
```
